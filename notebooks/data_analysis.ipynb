```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_swa = pd.read_csv('model_performance_metric_swa_variable_lr.csv')
df_swa.set_index('Epoch', inplace=True)
df_sgd = pd.read_csv('model_performance_metric_sgd.csv')
df_sgd.set_index('Epoch', inplace=True)
# plt.plot(df_sgd[['SGD_Model_Accuracy']])
plt.plot(df_swa['Running_Model_Accuracy'], label='Running Model Accuracy')
plt.plot(df_swa['SWA_Model_Accuracy'], label='SWA Model Accuracy')
plt.xlabel('Epochs')  # Modifier selon le contenu de votre DataFrame
plt.ylabel('Accuracy')
plt.title('Results of SGD and SWA on the test set accuracy')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('model_performance_metric_p_swa_cyclic_lr.csv')
df.set_index('Epoch', inplace=True)
plt.plot(df_swa['SWA_Model_Accuracy'], label='SWA Model Accuracy')
plt.plot(df[['Running_Model_Accuracy']], label = 'Running_Model_Accuracy')
plt.plot(df[['pSWA_Model_Accuracy']], label = 'pSWA_Model_Accuracy')

plt.xlabel('Epochs')  # Modifier selon le contenu de votre DataFrame
plt.ylabel('Accuracy')
plt.title('Results of different methods')
plt.legend()
plt.show()
```
